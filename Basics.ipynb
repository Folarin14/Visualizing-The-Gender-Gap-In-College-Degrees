```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
fig = plt.figure(figsize=(20,21))
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
cb_gray = (171/255,171/255,171/255)
for sp in range(1,18,3):
    cats_index = int(sp/3)
    ax1 = fig.add_subplot(6,3,sp)
    ax1.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cats_index]], c=cb_orange, label='Men', linewidth=3)
    ax1.plot(women_degrees['Year'], women_degrees[stem_cats[cats_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax1.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax1.set_yticks([0,100])
    ax1.axhline(50, c=cb_gray, alpha=0.3)
    for key, spine in ax1.spines.items():
        spine.set_visible(False)
    ax1.set_title(stem_cats[cats_index])
    ax1.set_xlim(1968,2011)
    ax1.set_ylim(0,100)
    #for axis in 
    if sp == 1:
        ax1.text(2008,87,'Men')
        ax1.text(2007,8,'Women')
    elif sp == 16:
        ax1.text(2007,62,'Men')
        ax1.text(2006,35,'Women')
        ax1.tick_params(labelbottom='on')
    
for sp in range(2,16,3):
    cats_index = int(sp/3)
    ax2 = fig.add_subplot(6,3,sp)
    ax2.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cats_index]], c=cb_orange, label='Men', linewidth=3)
    ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cats_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax2.tick_params(bottom='off', top='off', left='off', right='off',labelbottom='off')
    ax2.set_yticks([0,100])
    ax2.axhline(50, c=cb_gray, alpha=0.3)
    for key, spine in ax2.spines.items():
        spine.set_visible(False)
        ax2.set_title(lib_arts_cats[cats_index])
        ax2.set_xlim(1968,2011)
        ax2.set_ylim(0,100)
    if sp == 2:
        ax2.text(2005,75,'Women')
        ax2.text(2007,23,'Men')
    elif sp == 14:
        ax2.tick_params(labelbottom='on')
for sp in range(3,20,3):
    cats_index = int((sp-1)/3)
    ax3 = fig.add_subplot(6,3,sp)
    ax3.plot(women_degrees['Year'], 100-women_degrees[other_cats[cats_index]], c=cb_orange, label='Men', linewidth=3)
    ax3.plot(women_degrees['Year'], women_degrees[other_cats[cats_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax3.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax3.set_yticks([0,100])
    ax3.axhline(50, c=cb_gray, alpha=0.3)
    for key, spine in ax3.spines.items():
        spine.set_visible(False)
        ax3.set_title(other_cats[cats_index])
        ax3.set_xlim(1968,2011)
        ax3.set_ylim(0,100)
    if sp == 3:
        ax3.text(2006,88,'Women')
        ax3.text(2007,8,'Men')
    elif sp == 18:
        ax3.text(2007,62,'Men')
        ax3.text(2006,35,'Women')
        ax3.tick_params(labelbottom='on')
        
plt.show()  
    
```
